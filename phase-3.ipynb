```python
import numpy as np
```

```python
#data structure : [restrant id, 2021, 2022, 2023,2024]
import matplotlib.pyplot as plt

sales_data = np.array([[1, 150000, 180000, 220000, 250000],
                      [2, 120000, 140000, 160000, 190000],
                      [3, 200000, 230000, 260000, 300000],
                      [4, 180000, 210000, 240000, 270000],
                      [5, 160000, 185000, 205000, 230000],
                    ]
                      )

print("==== Zomato sales analyses ====")

print("\n Sales data shapes", sales_data.shape)

print("\n sample data for 1st 3 restorant: \n", sales_data[:3]) #same as 0:3
```

```python


#total sales per year
print(np.sum(sales_data, axis=0))
print(np.sum(sales_data, axis=1))


yearly_total = np.sum(sales_data[:, 1:], axis=0)
print("yeearly total", yearly_total)


# minimum sales  per restarant
min_sales = np.min(sales_data[:, 1:],axis=1)
print("min sales:", min_sales)

#max sales per year
max_sales = np.max(sales_data[:, 1:], axis=0)
print("max_sales", max_sales)


avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("avg sales", avg_sales)


cumsum = np.cumsum(sales_data[:, 1:], axis=1)
print("cumsum", cumsum)


plt.figure(figsize=(10, 6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Avrage cumulative sales  across all restorant")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
```
