```python
import numpy as np
arr_1d = np.array([1,2,3,4,6])
print("1d array", arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print("2d array", arr_2d)
```

#### list vs numpy array


```python
py_list = [1,2,3]
print("Python List Multiplication", py_list * 2)


np_array = np.array([1,2,3]) #element wise multiplications
print("Python List Multiplication", np_array * 2)

import time
start = time.time()
py_list = [i*2 for i in range(1000000)]
print("\n List operation time : ", time.time() - start)


start = time.time()
np_array = np.arange(1000000)
print("\n List operation time : ", time.time() - start)
```

###creating arrays from scratch


```python
zeros = np.zeros((3,4))

print("zeros array: \n", zeros)

once = np.ones((2,3))

print("once array: \n", once)

full = np.full((2,3), 7)

print("full array: \n", full)

random = np.random.random((2,3))


print("random array: \n", random)


seqence = np.arange(0, 11, 2)

print("seqence array: \n", seqence)
```

### Vector, Matrix and Tensor

```python
vector = np.array([1,2,3])
print("vector", vector)

matrix = np.array([[1,2,3],[3,4,5]])

print("matrix", matrix)

tensor = np.array([[[1,2],[3,4]], [[5,6],[7,8]]])
print("tensor", tensor)
```

### Array propertis

```python
arr = np.array([[1,2,3],[4,5,6]])

print("shapes of array", arr.shape)
print("Dimension", arr.ndim)
print("Size", arr.size)
print("Datatypee", arr.dtype)
```

### Array reshape


```python
ar = np.arange(0,12)
print("orignal array: \n", ar)

resahped = ar.reshape((3,4))
print(" \n reshaped array: \n", resahped)


flattened = resahped.flatten()
print('\n flattened array', flattened)

raveled = resahped.ravel()
# returned view, insted of copy
print("raveled array", raveled)


#transpose
transpose = resahped.T

print("transpose", transpose)
```
